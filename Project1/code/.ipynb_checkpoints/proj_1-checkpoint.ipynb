#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
final['act2017_participation'].dtype
````

````python
#standard deviation function
def my_std_dev(data):
    n = data.count()
    if data.dtype == float or data.dtype == int:
        mean_eng = data.mean()
    else:
        return 0
    square_sum = sum([(x-mean_eng)**2 for x in data])
    return round((square_sum/n)**0.5, 6)
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

````python
#Code:
sd = {i: my_std_dev(final[i]) for i in final.columns[1:]} #only to include numeric columns
sd
````

````python
#using np.std
{i: round(np.std(final[i]),6) for i in final.columns[1:]}
````

__Do any states show have >50% participation on both tests either year?__

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(13,8)) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column]) # feel free to add more settings
        # Set titles, labels, etc here for each subplot
        ax[i].set_title(list_of_titles[i])
        ax[i].set_xlabel(list_of_xlabels[i])
        
````

````python
def subplot_scatter(x, y, data, label_list, title):
#########
#     Input:
#     x = the x label in the dataset
#     y = the y label in the dataset
#     data = dataset
#     label = list of labels
#     title = title of your graph
#########
    #set the subplot size
    fig, ax = plt.subplots(figsize=(8,6))
    #plot scatterplot using seaborn with relevant details
    sns.scatterplot(x=x, y=y, data=data, axes=ax, s=130, alpha=1.0);
    
    #Set the look and feel
    ax.set_title(title, fontsize=20)
    ax.set_xlabel(label_list[0], fontsize=15)
    ax.set_ylabel(label_list[1], fontsize=15)
    plt.show()
````

#### Plot and interpret boxplots

For each numeric variable in the dataframe create a boxplot using Seaborn. Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

Each boxplot should:
- Only include variables of a similar scale
- Have clear labels for each variable
- Have appropriate titles and labels

````python
#Participation rate boxplot
ax = sns.boxplot(data=final[['act2017_participation', 'act2018_participation', 'sat2017_participation',\
                        'sat2018_participation']], orient='h', palette='Paired');
#set look and feel:
#set those ticklabels to make it more readable
ax.set_yticklabels(['ACT 2017', 'ACT 2018', 'SAT 2017', 'SAT 2018'], fontsize=14);
ax.set_title('Participation rate', fontsize=20);
#for tonedowning the colour 
# for patch in ax.artists:
#     r, g, b, a = patch.get_facecolor()
#     patch.set_facecolor((r, g, b, .3))
````

The participation rate of the ACT and SAT tests are quite different. The central tendency and spread are both dissimilar. For ACT, the data is more right skewed wheras for SAT it is more left skewed. The IQR and median is also different when comparing between the ACT and SAT rates. 

Between either years in each test, the ACT test has roughly the same spread while the SAT test has a slightly different spread. The SAT 2017 test has a shorter IQR box and lower median as compared to the SAT 2018 test. 

````python
def subplot_box(data, xticklabels, title, palette='Blues', figsize=(7,5)):
#########
#     Input:
#     data = dataset
#     xticklabels = the x tick labels you want in each boxplot
#     title = title of your graph
#     palette = specify any palette, default will be Blues
#     figsize = specify how big you want the plot to be, default is (7,5)
#########
    fig, ax = plt.subplots(figsize=figsize);
    sns.boxplot(data=data,ax=ax, palette=palette); #since we are not highlighting any boxplot will default it to Blues
    #set look and feel:
    #set those ticklabels to make it more readable
    ax.set_xticklabels(xticklabels, fontsize=14);
    ax.set_title(title, fontsize=20);
````
